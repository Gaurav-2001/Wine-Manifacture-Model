```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import RMSprop
```

```python
ds=pd.read_csv('wines.csv')
ds.head()
```

```python
ds.columns
```

```python
y=ds['Class']
y.value_counts()
```

```python
y_cat=pd.get_dummies(y)
```

```python
X=ds.drop('Class',axis=1)
X.columns
```

```python
model=Sequential()
```

```python
model.add(
    Dense(
        units=5,
        input_shape=(13,),
        activation='relu',
        kernel_initializer='he_normal'
    )
)

model.add(
    Dense(
        units=8,
        activation='relu',
        kernel_initializer='he_normal'
    )
)

model.add(
    Dense(
        units=8,
        activation='relu',
        kernel_initializer='he_normal'
    )
)

model.add(
    Dense(
        units=3,
        activation='softmax'
    )
)
```

```python
model.summary()
```

```python
model.compile(
    optimizer=RMSprop(learning_rate=0.001),
    loss = 'categorical_crossentropy',
    metrics=['accuracy'],
)
```

```python
model.fit(X,y_cat,epochs=1000)
```

```python
l=pd.DataFrame(model.history.history['accuracy'])
l.plot()
```

```python
model.save('Wine.h5')
```
